```python
import numpy as np
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import re, time,csv
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import platform
import os
```

```python
"""
Extracts all the story links from the current page and writes to the specified writer
"""
def get_links_from_page( writer:csv.writer ):
    results=driver.find_elements(by=By.CSS_SELECTOR,value='div[data-hook="review"]')
    for result in results:

        name,ratings,date,review='NA','NA','NA','NA'
       
        try:
            name_box=result.find_element(by=By.CSS_SELECTOR,value='span[class="a-profile-name"]')                                                                      
            name=name_box.text
            #print(name)
        except NoSuchElementException as e: 
            print('could not extract name')
            
        try:
            ratings_box=result.find_element(by=By.CSS_SELECTOR,value='i[data-hook="review-star-rating"]')
            ratings= re.search(r'\d+', ratings_box.get_attribute('class')).group()
            #print(ratings)
        except NoSuchElementException as e:
            print('could not extract title')

        try:
            text_box=result.find_element(by=By.CSS_SELECTOR,value='div[class="a-row a-spacing-small review-data"]')
            review_box=text_box.find_element(by=By.CSS_SELECTOR,value='span[class="cr-translated-review-content"]')
            review=review_box.text
            #print(review)
        except NoSuchElementException as e:
            try:
                text_box=text_box.find_element(by=By.CSS_SELECTOR,value='span[data-hook="review-body"]')
                review=text_box.text
                #print(text_box.text)

            except NoSuchElementException as e:
                print('could not extract text')

        try: 
            date_box=result.find_element(by=By.CSS_SELECTOR,value='span[data-hook="review-date"]')
            date_pattern = r'[A-Z][a-z]+\s\d{1,2},\s\d{4}'
            # Use regular expression to find the date pattern in the text
            date = re.search(date_pattern, date_box.text).group()
            #print(date)
        except NoSuchElementException as e:
            print('could not extract date')
            

        writer.writerow([name,ratings,review,date])
```

```python

def scrape_query(query:str, url:str, delay:int = 5):
    
    file_name = query+'_amazon_reviews.csv'

    fw=open(file_name,'w',encoding="utf-8")
    writer=csv.writer(fw,lineterminator='\n')
    writer.writerow(['name','ratings','review','date'])
    driver.get(url)

    driver.refresh()
    time.sleep(0.5)  # Short pause between refreshes; adjust as needed
    # driver.refresh()
     
#     url='https://www.amazon.com/Productivity-Qualcomm-Snapdragon-Included-Graphite/product-reviews/B09NQMJ5XG/ref=cm_cr_getr_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&sortBy=recent&pageNumber='
    for star_rating in reversed(range(1, 6)):
        page_cnt=1 
        time.sleep(2)
        try:
            dropdown_button = driver.find_element(by=By.ID, value='a-autoid-5-announce')
            driver.execute_script("arguments[0].click();", dropdown_button)
            time.sleep(2)

            star_option_id = f"star-count-dropdown_{6 - star_rating}"  
            star_option_xpath = f"//a[@id='{star_option_id}']"
            star_option = driver.find_element(by=By.XPATH, value=star_option_xpath)
            driver.execute_script("arguments[0].click();", star_option)
            time.sleep(delay)
        except NoSuchElementException:
            print(f'Could not select {star_rating} star rating')
            continue
            
        time.sleep(delay)
        while(True): 
             
            print('page',page_cnt) 
            # s=url+str(page_cnt)
            # driver.get(s) 
            time.sleep(delay)
            
            try:
                translate=driver.find_element(by=By.CSS_SELECTOR,value='a[data-hook="cr-translate-these-reviews-link"]')
                driver.execute_script("arguments[0].click();", translate)
            except NoSuchElementException as e:
                print("Translated")

            
            time.sleep(3)
            get_links_from_page(writer) 
            
            WebDriverWait(driver, delay)
            
            try:
                
                next_page_xpath = "//li[@class='a-last']/a"
                next_page_link = driver.find_element(by=By.XPATH, value=next_page_xpath)
                driver.execute_script("arguments[0].click();", next_page_link)
                time.sleep(delay)  # Wait for the page to load

            except NoSuchElementException:
                # print("Next page link not found, assuming last page reached")
                break
            
            page_cnt+=1 
            time.sleep(delay)

    fw.close()
```
